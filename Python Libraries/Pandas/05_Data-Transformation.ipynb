```python
"""Sorting & Ranking"""
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.sort_values("Year")
```

```python
df.sort_values('Year', ascending = False) #for decending sort
```

```python
df.sort_values(["Year", "IMDb"], ascending = False )
```

```python
df2 = df.sort_values(["Year", "IMDb"], ascending = [True, False] )
df2
```

```python
df2.reset_index() # to sort the index and also return old index in new column named index
```

```python
df2.reset_index(drop = True) # Using drop will not add new column of old index
```

```python
df2.reset_index(drop = True, inplace = True) # By using 'inplace' we can update the object
```

```python
df2
```

```python
df3 = df.sort_values("BoxOffice(INR Crore)", ascending = False).copy()
df3
```

```python
df3.sort_index() 
```

```python
df["Rank"] = df["IMDb"].rank() # Creates a new column with ranking of data based on IMDb in ascending order.
df
```

```python
df["Rank"] = df["IMDb"].rank(ascending = False, method = "dense") # method = 'dense' will hwlp to return ranking in int or whole no.
df
```

```python
df.rename(columns={"Film":"Movie Name"}, inplace = True) # if we not use inplace = True then we will receive a view 
df
```

```python
df.columns = ["Actor", "Film", "Year", "Type", "BOxOffice(INR Crore)", "IMDb", "Rank"] # Used to change the name of all columns
df
```

```python
df = df[["Film", "Year", "IMDb", "Actor","BOxOffice(INR Crore)", "Type" ]]#reorder columns
df
```
