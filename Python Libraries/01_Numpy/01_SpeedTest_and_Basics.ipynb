```python
#Using Python Lists
import time

size = 10_000_000

list_1 = list(range(size))
list_2 = list(range(size))

start = time.time()
list_3 = [x+y for x,y in zip(list_1, list_2)]
end = time.time()

print(end-start)
```

```python
#Using Python Lists
import numpy as np
import time

size = 10_000_000

list_1 = np.array(range(size))
list_2 = np.array(range(size))

start = time.time()
list_3 = list_1 + list_2 #Vectorized operation
end = time.time()

print(end-start)
```
