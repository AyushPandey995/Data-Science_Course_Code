```python

""" 'json.dumps()' is used to convert python object to jsondata/string."""
import json
obj = {"Name" : "Ayush", "Course" : "BE", "Branch" : "Computer Engineering", "Year" : 4}
json_string = json.dumps(obj)
print(type(obj), obj)
print(type(json_string), json_string)
```

```python

""" 'json.loads()' is used to convert string/jsondata to its actual from."""
json_string1 = '{"Name" : "Ayush", "Course" : "BE", "Branch" : "Computer Engineering", "Year" : 4}'
python_obj = json.loads(json_string1)
print(type(python_obj), python_obj)
```

```python

"""'json.dump()' is used to write in any json file/ add content to json file"""

with open("sample.json", "w") as f:
    json.dump(obj, f)
```

```python

"""'json.load()' is used to read/print the content of json file and helps to store data in an object"""
with open("sample.json", "r") as f:
    python_data = json.load(f)
    
print(python_data)
print(type(python_data))
```

```python

""" indent is used to beautify the string by adding spaces"""
obj = {"Name" : "Ayush", "Course" : "BE", "Branch" : "Computer Engineering", "Year" : 4}
json_string = json.dumps(obj, indent = 4)
print(type(obj), obj)
print(type(json_string), json_string)
```
