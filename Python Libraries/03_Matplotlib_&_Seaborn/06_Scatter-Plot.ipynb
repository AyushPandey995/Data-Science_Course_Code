```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
plt.scatter(study_hours, exam_scores)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Marks Scored based on Study Hours")
plt.figure(figsize = (8, 10))
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 85, 52, 65, 75, 85, 90]
colors = ['red' if score < 60 else 'green' for score in exam_scores]
sizes = [score*1.5 for score in exam_scores]
plt.scatter(study_hours, exam_scores, c=colors, s=sizes)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Marks Scored based on Study Hours")
plt.figure(figsize = (8, 10))
```

```python
'''cmap adds gradient coloring based on the score. colorbar helps understand what the colors represent.

'''
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 85, 52, 65, 75, 85, 90]
sizes = [score*1.5 for score in exam_scores]
plt.scatter(study_hours, exam_scores, s=sizes, c=exam_scores, cmap = 'viridis')
plt.grid(True)
plt.colorbar(label = 'Score') # Use to add colorbar on side of plot
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Marks Scored based on Study Hours")
plt.figure(figsize = (8, 10))
plt.tight_layout()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 85, 52, 65, 75, 85, 90]
behaviour_scores = [50, 40, 80, 60, 85, 70, 90, 88, 65]
sizes = [score*1.5 for score in exam_scores]
plt.scatter(study_hours, exam_scores, s=sizes, c=behaviour_scores, cmap = 'viridis')
plt.grid(True)
plt.colorbar(label = 'Behaviour Score') # Use to add colorbar on side of plot
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Marks Scored based on Study Hours")
plt.figure(figsize = (8, 10))
plt.tight_layout()
```

```python
'''Annotations help identify individual points, which is useful in small datasets.

'''
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 85, 52, 65, 75, 85, 90]
behaviour_scores = [50, 40, 80, 60, 85, 70, 90, 88, 65]
sizes = [score*1.5 for score in exam_scores]
plt.scatter(study_hours, exam_scores, s=sizes, c=behaviour_scores, cmap = 'viridis')
for i in range(len(study_hours)):
    plt.annotate(f"S{i+1}", (study_hours[i]+0.1, exam_scores[i]-1.5))
plt.grid(True)
plt.colorbar(label = 'Behaviour Score') # Use to add colorbar on side of plot
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Marks Scored based on Study Hours")
plt.figure(figsize = (8, 10))
plt.tight_layout()
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
plt.scatter(class_a_hours, class_a_scores, label = "A", color = "blue")
plt.scatter(class_b_hours, class_b_scores, label = "B", color = "red")
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Marks Scored based on Study Hours")
plt.legend()
```
