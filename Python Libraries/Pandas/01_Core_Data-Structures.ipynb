# Getting Started with Pandas
## What is Pandas?
Pandas is a powerful, open-source Python library used for data manipulation, cleaning, and analysis. It provides two main data structures:

- Series: A one-dimensional labeled array
- DataFrame: A two-dimensional labeled table (like an Excel sheet or SQL table)
- Pandas makes working with structured data fast, expressive, and flexible.

## Pandas vs Excel vs SQL vs NumPy
--------------
#### Tool------------Strengths--------------------Weaknesses<br>
Excel-------Easy UI, great for small data--------Slow, manual, not scalable<br>
SQL---------Efficient querying of big data-------Not ideal for transformation logic<br>
NumPy-------Fast, low-level array operations-----No labels, harder for tabular data<br>
Pandas------Label-aware, fast, flexible----------Slightly steep learning curve<br>

Pandas is built on two main data structures:
Series → One-dimensional (like a single column in Excel)
DataFrame → Two-dimensional (like a full spreadsheet or SQL table)

1. Series — 1D Labeled Array.
A Series is like a list with labels (index).

2. DataFrame — 2D Labeled Table.
A DataFrame is like a dictionary of Series — multiple columns with labels.

Most Pandas operations are built on these foundations:

- Selection
- Filtering
- Merging
- Aggregation<br>
Understanding Series & DataFrames will make everything else easier.



```python
import pandas as pd
```

```python
s1 = pd.Series([93, 74, 88, 71, 96, 68, 77, 91])
print(type(s1))
print(s1)
```

```python
s2 = pd.Series([93, 74, 88, 71, 96, 68, 77, 91], index = ["Ayush", "Vainhav", "Shail", "Praveen", "Khushi", "Darshan", "Shreya", "Anush"])
print(type(s2))
print(s2)
```

```python
data = {
    "Roll_No" : [101, 102, 103, 104, 105, 106],
    "Name" : ["Ayush", "Kanha", "Khushi", "Lucky", "Omi", "Shraddha"],
    "Marks" : [93, 91, 92, 95, 97, 94]
}
df = pd.DataFrame(data)
print(type(df))
print( "\n", df)
```

```python
print(df.index)
print(df.columns)
```

```python
df.index = ["a", "b", "c", "d", "e", "f"]
print(df)
```

```python
df.columns = ["Seat_No", "First_Name", "Avg_Marks"]
print(df)
```
