```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Sachine Tendulkar's Yearly Runs")
plt.plot(years, runs)
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kolhi")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
plt.title("Runs Comparison")
```

```python
plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label="Sehwag")  # green triangles dotted
plt.legend()
```

```python
Sachin =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, Sachin, color='blue', label="Tendulkar")
plt.legend()
```

```python
#Color and Line Style Arguments
plt.plot(years, kohli, color = "Red", linestyle = "--", marker = "o", linewidth = 3, label = "Virat" )
plt.plot(years, Sachin, color = "Blue", linestyle = "-.", marker = "p", linewidth = 2, label = "Sachin" )
plt.xlabel = "Year"
plt.ylabel = "Runs"
plt.legend()
```

```python
plt.plot(years, kohli, color = "Red", linestyle = "--", marker = "o", linewidth = 3, label = "Virat" )
plt.plot(years, Sachin, color = "Blue", linestyle = "-.", marker = "p", linewidth = 2, label = "Sachin" )
plt.xlabel = "Year"
plt.ylabel = "Runs"
plt.grid(True)
plt.legend()
```

```python
plt.plot(years, kohli, color = "Red", linestyle = "--", marker = "o", linewidth = 3, label = "Virat" )
plt.plot(years, Sachin, color = "Blue", linestyle = "-.", marker = "p", linewidth = 2, label = "Sachin" )
plt.xlabel = "Year"
plt.ylabel = "Runs"
plt.tight_layout()
plt.grid(True)
plt.legend()
```
