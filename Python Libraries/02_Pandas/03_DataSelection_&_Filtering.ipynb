```python
'''Querying with .query()'''
print("Output 1")
print(df.query("IMDb > 8"))

print("\n\nOutput 2")
print( df.query('IMDb > 7 and Year > 2020'))#and

print("\n\nOutput 3")
print( df.query('IMDb > 8 or Year > 2020'))#or

'''String values must be in quotes'''
print("\n\nOutput 4")
print( df.query('9 > IMDb > 6 and Genre == "Comedy" '))

'''Use backticks for column names with spaces or special characters
If a column name has spaces, use backticks (`):'''
#Ex - df.query("`first name` == 'Alice'")

''' You can use @ to reference Python variables'''
print("\n\nOutput 5")
year = 2018
print(df.query("Year > @year"))
```
