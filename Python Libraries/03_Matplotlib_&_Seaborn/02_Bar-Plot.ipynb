```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
 
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
 
plt.plot(years, sachin) #Line plot
```

```python
plt.bar(years, sachin)
plt.xlabel("Years")
plt.xlabel("Runs")
plt.title("Sachin Tendulkar's Yearly runs")
```

```python
plt.bar(years, sachin, color = "blue", label = "Sachin")
plt.bar(years, kohli, color = "red", label = "Virat")
plt.bar(years, sehwag, color = "green", label = "Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
plt.title("Runs Comparison")
```

```python
#Here numpy arange function helps to arrange the bars of multiple players side by side

import numpy as np
x = np.arange(len(years))
width = 0.25


plt.bar(x-width, sachin, width = width, color = "blue", label = "Sachin")
plt.bar(x, kohli, width = width, color = "red", label = "Virat")
plt.bar(x+width, sehwag, width = width, color = "green", label = "Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
plt.xticks(x, years)
plt.tight_layout()
plt.title("Runs Comparison")
```

```python
#Horizontal bar plot
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players, runs_5yrs, color = "pink")
plt.xlabel("First 5 year Performance of Indian Batsman")
```

```python
#Add text at the top of each bar.
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color = "pink")
plt.xlabel("Performance of Indian Batsman")
plt.ylabel("Runs")
for i in range(len(players)):
    plt.text(i, runs[i]+20, str(runs[i]), ha = 'center')
```

```python

"""Assignment - Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
Then create a horizontal bar chart showing total runs scored in their debut 5 years.
"""
import numpy as np

years = [2003, 2006, 2009, 2011, 2013]

sachin = [1141, 623, 972, 513, 287]
kohli  = [0, 0, 1073, 1381, 1268]
sehwag = [871, 608, 810, 645, 31]

x = np.arange(len(years))
width = 0.25

plt.bar(x-width, sachin, width = width, color = "Blue", label = "Sachin") 
plt.bar(x, kohli, width = width, color = "Red", label = "Kohli") 
plt.bar(x+width, sehwag, width = width, color = "Green", label = "Sehwag") 

# Sachin highest
idx = sachin.index(max(sachin))
plt.text(x[idx] - width, max(sachin) + 30,
         str(max(sachin)), ha='center', color='blue', fontweight='bold')

# Kohli highest
idx = kohli.index(max(kohli))
plt.text(x[idx], max(kohli) + 30,
         str(max(kohli)), ha='center', color='red', fontweight='bold')

# Sehwag highest
idx = sehwag.index(max(sehwag))
plt.text(x[idx] + width, max(sehwag) + 30,
         str(max(sehwag)), ha='center', color='green', fontweight='bold')


plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Comparison")
plt.xticks(x, years)
```
