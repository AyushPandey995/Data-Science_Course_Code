```python
sns.lineplot(x = 'total_bill', y = 'tip', data = data)
plt.title("Tip analysis based on total bill")
plt.xlabel("Bill Paid")
plt.ylabel("Given Tip")
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = data, hue="time")
plt.title("Tip analysis based on total bill")
plt.xlabel("Bill Paid")
plt.ylabel("Given Tip")
```

```python
sns.barplot(x = 'total_bill', y = 'tip', data = data)
plt.title("Tip analysis based on total bill")
plt.xlabel("Bill Paid")
plt.ylabel("Given Tip")
```

```python
sns.histplot(x = 'total_bill', y = 'tip', data = data)
plt.title("Tip analysis based on total bill")
plt.xlabel("Bill Paid")
plt.ylabel("Given Tip")
```

```python
sns.boxplot(x = 'total_bill', y = 'tip', data = data)
plt.title("Tip analysis based on total bill")
plt.xlabel("Bill Paid")
plt.ylabel("Given Tip")
```
