```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, ]
y1 = [x*2 for x in x]
y2 = [x**2 for x in x]

plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
```

```python
x = [1, 2, 3, 4, 5, ]
y1 = [x*2 for x in x]
y2 = [x**2 for x in x]

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.xlabel("Numbers")
plt.ylabel("Double of Numbers")
plt.grid(True)
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.xlabel("Numbers")
plt.ylabel("Square of Numbers")

plt.grid(True)
plt.tight_layout()
```

```python
x = [1, 2, 3, 4, 5, ]
y1 = [x*2 for x in x]
y2 = [x**2 for x in x]
y3 = [x**1/2 for x in x]
y4 = [x/2 for x in x]

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.xlabel("Numbers")
plt.ylabel("Double of Numbers")
plt.grid(True)
plt.tight_layout()

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.xlabel("Numbers")
plt.ylabel("Square of Numbers")
plt.grid(True)
plt.tight_layout()

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.xlabel("Numbers")
plt.ylabel("Squareroot of Numbers")
plt.grid(True)
plt.tight_layout()

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.xlabel("Numbers")
plt.ylabel("Half of Numbers")
plt.grid(True)
plt.tight_layout()
```

```python

fig, axs=plt.subplots(1, 2, figsize=(10, 6))
axs[0].plot(x, y1)
axs[0].set_title("x*2")

axs[1].plot(x, y2)
axs[1].set_title("x**2")
```

```python

fig, axs=plt.subplots(2, 2, figsize=(6, 8))
axs[0, 0].plot(x, y1)
axs[0, 0].set_title("x*2")

axs[0, 1].plot(x, y2)
axs[0, 1].set_title("x**2")

axs[1, 0].plot(x, y3)
axs[1, 0].set_title("x**1/2")

axs[1, 1].plot(x, y4)
axs[1, 1].set_title("x/2")

fig.suptitle("Comparison Plot", fontsize=16)
fig.tight_layout()
```
