```python
import pandas as pd
```

```python
"""Mergeing"""
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

employees
```

```python
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})

departments
```

```python
pd.merge(employees, departments, on= "DeptID")
```

```python
pd.merge(employees, departments, on= "DeptID", how="left")
```

```python
pd.merge(employees, departments, on = "DeptID", how = "right")
```

```python
pd.merge(employees, departments, on = "DeptID", how = "outer")
```

```python
'''Concatenating DataFrames - Joining'''

df1 = pd.DataFrame({"Name": ["Alice", "Bob"]})
df1
```

```python
df2 = pd.DataFrame({"Name": ["Charlie", "David"]})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
df1 = pd.DataFrame({"Name": ["Alice", "Bob"], "Score" : [556, 784]})
df1
```

```python
df2 = pd.DataFrame({"Name": ["Charlie", "David"], "Age" : [45, 78]})
df2
```

```python
pd.concat([df1, df2])
```
