```python
import matplotlib.pyplot as plt
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels)
plt.title("Run comparison of Indian Players")
```

### Parameter -------- Use<br>
labels -------- Label each slice<br>
colors -------- Customize slice colors (names or hex)<br>
explode -------- Pull out slices for emphasis<br>
shadow -------- Adds depth-like shadow<br>
startangle -------- Rotates pie to start from a different angle<br>
autopct -------- Shows percentage text on slices<br>
wedgeprops -------- Customize slice edge, fill, width, etc.<br>

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors)
plt.title("Run comparison of Indian Players")
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors,
       wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1.2, 'linestyle' : "--" })
plt.title("Run comparison of Indian Players")
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors,
       wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1.2, 'linestyle' : "--", 'fill' : False })
plt.title("Run comparison of Indian Players")
```

```python
explode = [0.2, 0, 0, 0]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors,
       wedgeprops = {'edgecolor' : 'black', 'linewidth' : 1.2, 'linestyle' : "--" },
       explode = explode)
plt.title("Run comparison of Indian Players")
```

```python
explode = [0.2, 0, 0, 0]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors, explode = explode, shadow = True)
plt.title("Run comparison of Indian Players")
```

```python
explode = [0.2, 0, 0, 0]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title("Run comparison of Indian Players")
```

```python
explode = [0.2, 0, 0, 0]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs, labels = labels, colors = colors, explode = explode, shadow = True, autopct = '%1.1f%%')
plt.title("Run comparison of Indian Players")
```

```python
"""Create a pie chart showing the market share of mobile OS (Android, iOS, others). 
Then recreate the same using a horizontal bar chart and observe which is easier to understand.
"""
explode = [0.1, 0, 0]
os_names = ["Android", "iOS", "Others"]
market_share = [72, 27, 1]

plt.pie(market_share, labels = os_names, explode = explode, shadow = True, autopct = '%1.1f%%')
```

```python
plt.barh(os_names, market_share, color = "pink")
plt.xlabel("Percentage of use")
plt.ylabel("Category")
plt.tight_layout()
```
