```python
import json
```

```python
def load_data(filename):
    with open (filename, "r") as f:
        return json.load(f)
    
```

```python
def users_detail(data):
    for users in data['users']:
        print("\n\nName and their Connections-")
        print(f"Id & Name - {users['id']} : {users['name']}")
        print("Friends- ")
        print("ID : Name")
        for friends_id in users['friends']:
            for u in data['users']:
                if u['id'] == friends_id:
                    print(f"{u['id']} : {u['name']}")
        for page in users['liked_pages']:
            for page_id in data['pages']:
                if page_id['id'] == page:
                    print(f"{page_id['id']} : {page_id['name']}")
            

    print("\n\nPage ID & their Name")
    for page in data['pages']:
        print(f"{page['id']} : {page['name']}")
```

```python
users_detail(content)
```
